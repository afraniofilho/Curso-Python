```python
# criar um novo dataframe sem os registros que contêm a palavra Ralançamento
df_novo = df[df['Data de lançamento'] != 'Relançamento'].copy()
```

```python
df_novo.shape
```

```python
# Quantos registros existem com relaçamento
df_novo[df_novo['Data de lançamento'] == 'relançamento']
```

```python
# retirar o registro com o dado relançamento do df_novo
df_novo.drop(index=531, inplace=True)
```

```python
# converter dados da coluna Data de lançamento em dados do time datetime
df_novo['Data de lançamento'] = pd.to_datetime(df_novo['Data de lançamento'])
```

```python
df_novo.info()
```

```python
# Quais anos tiveram mais lançamentos?
df_novo['Data de lançamento'].dt.year.value_counts()
```

```python
df_novo_dt = df_novo[df_novo['Data de lançamento'].dt.year > 2008].copy()
```

```python
df_novo_dt[df_novo_dt['Data de lançamento'].dt.year < 2008].count()
```

```python
# Qual o filme com maior renda
df_novo[df_novo['Renda (R$) no ano de exibição'] == df_novo['Renda (R$) no ano de exibição'].max()]
```
