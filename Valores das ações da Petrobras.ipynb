```python
# importar biblioteca datetime
import datetime
```

```python
#  importar biblioteca gráfica
import matplotlib.pyplot as plt
```

```python
# data inicial da busca
inicio = datetime.datetime(2001, 1, 1)
```

```python
# data final da busca
final = datetime.datetime(2021, 2, 1)
```

```python
inicio, final
```

```python
# Requisitar os dados no site do yahoo e guardar na variável resposta
resposta = web.DataReader('VALE3.SA', 'yahoo', inicio, final)
```

```python
# ler as 5 primeiras linhas
resposta.head()
```

```python
resposta.tail()
```

```python
resposta.shape
```

```python
# Gerar um gráfico com os valores das cotações
%matplotlib notebook

plt.plot(resposta.Close)
plt.plot(resposta.Open)
plt.title('Ações Petrobras de 2001 a 2021')
plt.xlabel('Data')
plt.ylabel('$')
plt.show()
```

```python
# qual a data e valor da maior cotação no período
resposta.loc[resposta['Close'] == resposta['Close'].max()]
```
